```python
import tkinter as tk
import numpy as np
import cv2
from googletrans import Translator

#pip install pillow
from PIL import ImageTk,Image,ImageDraw

from keras.models import Sequential
from keras.layers import Dense,Flatten

model=Sequential()

model.add(Flatten(input_shape=(28,28)))

model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(10,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.load_weights('FFNN-MNIST.h5')
num2words1 = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four', 5: 'Five', \
            6: 'Six', 7: 'Seven', 8: 'Eight', 9: 'Nine', 10: 'Ten', \
            11: 'Eleven', 12: 'Twelve', 13: 'Thirteen', 14: 'Fourteen', \
            15: 'Fifteen', 16: 'Sixteen', 17: 'Seventeen', 18: 'Eighteen', 19: 'Nineteen'}
num2words2 = ['Twenty', 'Thirty', 'Forty', 'Fifty', 'Sixty', 'Seventy', 'Eighty', 'Ninety']




def number(Number):
    if 0 <= Number <= 19:
        return num2words1[Number]
    elif 20 <= Number <= 99:
        tens, remainder = divmod(Number, 10)
        return num2words2[tens - 2] + ' ' + num2words1[remainder] if remainder else num2words2[tens - 2]
    else:
        print('Number out of implemented range of numbers.')

def event_function(event):
    
    x=event.x
    y=event.y
    
    x1=x-30
    y1=y-30
    
    x2=x+30
    y2=y+30

    canvas.create_oval((x1,y1,x2,y2),fill='black')
    img_draw.ellipse((x1,y1,x2,y2),fill='white')

def save():
    
    global count
    
    img_array=np.array(img)
    img_array=cv2.resize(img_array,(28,28))
    
    cv2.imwrite(str(count)+'.jpg',img_array)
    count=count+1
    
def clear():
    
    global img, img_draw
    
    canvas.delete('all')
    img=Image.new('RGB',(500,500),(0,0,0))
    img_graw=ImageDraw.Draw(img)
    
    label_status.config(text='PREDICTED DIGIT: NONE')
    

def add():
    global a,img,img_draw
    
    img_array=np.array(img)
    img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2GRAY)
    img_array=cv2.resize(img_array,(28,28))
    
    img_array=img_array/255.0
    img_array=img_array.reshape(1,28,28)
    result=model.predict(img_array)
    label=np.argmax(result,axis=1)
    
    
    print(int(label))
    a = a + int(label)
    
    
    canvas.delete('all')
    img=Image.new('RGB',(500,500),(0,0,0))
    img_draw=ImageDraw.Draw(img)    
    
    label_status.config(text='PREDICTED DIGIT:'+str(label))
    
    
    
def answer():
    add()
    win1 = tk.Tk()
    label_status1=tk.Label(win1,text='Sum',font='Helvetica 24',justify='left')
    label_status1.grid(row=2,column=0,columnspan=4)
    print("\033[1m"+"SUM = "+"\033[0m",end="")
    print(a)
    print("\033[1m",end="")
    print(number(a))
    print("\033[0m")
    win.destroy()
    translator = Translator()
    result_hindi = translator.translate(number(a),dest='hi')
    result_german = translator.translate(number(a),dest='de')

    result_french = translator.translate(number(a),dest='fr')
    result_russian = translator.translate(number(a),dest='ru')
    result_arabic = translator.translate(number(a),dest='ar')
    result_japanese = translator.translate(number(a),dest='ja')
    result_portuguese = translator.translate(number(a),dest='pt')
    result_chinese = translator.translate(number(a),dest='zh-cn')
    result_korean = translator.translate(number(a),dest='ko')
    print("\nHindi = "+result_hindi.text)
    print("\nGerman = "+result_german.text)
    print("\nFrench = "+result_french.text)
    print("\nRussian = "+result_russian.text)
    print("\nArabic = "+result_arabic.text)
    print("\nJapanese = "+result_japanese.text)
    print("\nPortuguese = "+result_portuguese.text)
    print("\nChinese = "+result_chinese.text)
    print("\nKorean = "+result_korean.text)
    label_status1.config(text='SUM = '+str(a)+'\nIn Words = '+number(a)+"\nHindi = "+result_hindi.text
                        +"\nGerman = "+result_german.text+"\nFrench = "+result_french.text
                        +"\nRussian = "+result_russian.text+"\nArabic = "+result_arabic.text
                        +"\nJapanese = "+result_japanese.text+"\nPortuguese = "+result_portuguese.text
                        +"\nChinese = "+result_chinese.text+"\nKorean = "+result_korean.text)
    
count=0
a=0

    
win=tk.Tk()


canvas=tk.Canvas(win,width=500,height=500,bg='white')
canvas.grid(row=0,column=0,columnspan=4)

button_save=tk.Button(win,text='SAVE',bg='green',fg='white',font='Helvetica 20 bold',command=save)
button_save.grid(row=1,column=0)

button_add=tk.Button(win,text='ADD',bg='yellow',fg='white',font='Helvetica 20 bold',command=add)
button_add.grid(row=1,column=1)

button_exit=tk.Button(win,text='ANSWER',bg='red',fg='white',font='Helvetica 20 bold',command=answer)
button_exit.grid(row=1,column=2)

button_clear=tk.Button(win,text='CLEAR',bg='blue',fg='white',font='Helvetica 20 bold',command=clear)
button_clear.grid(row=1,column=3)

label_status=tk.Label(win,text='PREDICTED DIGIT: NONE',bg='white',font='Helvetica 24 bold')
label_status.grid(row=2,column=0,columnspan=4)



canvas.bind('<B1-Motion>',event_function)
img=Image.new('RGB',(500,500),(0,0,0))
img_draw=ImageDraw.Draw(img)


win.mainloop()
```
